#### let's predict 'postPSS' based on 'prePSS'.



```python
import statsmodels.api as sm

# Adding a constant for the intercept
X = sm.add_constant(df['prePSS'])
y = df['postPSS']

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression results
print(model.summary())
```
